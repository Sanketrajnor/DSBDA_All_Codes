```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airquality.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
subset=df.loc[1:10,['Date','PM2.5','AQI']]
subset
```

```python
x = subset["Date"]
y = subset["AQI"]
```

```python
plt.xticks(np.arange(len(x)), x, rotation=90)
plt.title('Datewise Air Quality Index')
plt.plot(x, y)
plt.show()
```

```python
plt.bar(x,y)
plt.xticks(np.arange(len(x)), x, rotation=90)
plt.title('Datewise Air Quality Index')
plt.show()
```

```python
plt.scatter(x,y)
plt.xticks(np.arange(len(x)), x, rotation=90)
plt.title('Datewise Air Quality Index')
plt.show()
```

```python
# average_pm25 = df["PM2.5"].mean()
# average_aqi = df["AQI"].mean()
# average_pm10 = df["PM10"].mean()
average_no = df["NO"].count()
average_no2 = df["NO2"].count()
average_nox = df["NOx"].count()
average_nh3 = df["NH3"].count()
average_co = df["CO"].count()
average_so2 = df["SO2"].count()
average_o3 = df["O3"].count()
average_benzene = df["Benzene"].count()
average_toluene = df["Toluene"].count()
average_xylene = df["Xylene"].count()
labels = ["NO","NO2","NOx","NH3","CO","SO2","O3","Benzene","Toluene","Xylene"]
values = [average_no,average_no2,average_nox,average_nh3,average_co,average_so2,average_o3,average_benzene,average_toluene,average_xylene]
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Average Values")
plt.show()
```

```python
subset_sorted = subset.sort_values(by='Date')
x_asc = subset_sorted['Date']
y_asc = subset_sorted['AQI']
plt.step(x_asc, y_asc,linewidth=2)
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('Staircase Plot of AQI over Time')
plt.xticks(rotation=90)
plt.show()
```
