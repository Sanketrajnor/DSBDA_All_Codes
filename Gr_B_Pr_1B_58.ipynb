```python
data = {
    'name of person': ['Sanket', 'Abhishek', 'Darshan', 'Sumit', 'Kamlesh'],
    'height': [175, 54, 180, 5.5, 190],
    'unit': ['cm', 'inch', 'cm', 'feet', 'cm']
}

dataf = pd.DataFrame(data)

print("Original DataFrame:\n",dataf)

dataf.loc[dataf['unit'] == 'inch', 'height'] *= 2.54
dataf.loc[dataf['unit'] == 'feet', 'height'] *= 30.48
dataf['unit'] = 'cm'

print("\nCorrected DataFrame:\n",dataf)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
from sklearn.metrics import classification_report
```

```python
x = ht[['cp','thalachh','exng','oldpeak','slp','caa']]

y = ht['output']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
print(f"Size of splitted data")
print(f"x_train {x_train.shape}")
print(f"y_train {y_train.shape}")
print(f"x_text {x_test.shape}")
print(f"y_test {y_test.shape}")
```

```python
#Naive Baye's Classifier
from sklearn.naive_bayes import GaussianNB
model_hd = GaussianNB()
model_hd.fit(x_train, y_train)
y_pred_hd = model_hd.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_hd)
plt.title('Naive Bayes')
plt.show()
print(f" Accuracy is {accuracy_score(y_test, y_pred_hd)}")
print(classification_report(y_test, y_pred_hd))#Naive Baye's Classifier
```
