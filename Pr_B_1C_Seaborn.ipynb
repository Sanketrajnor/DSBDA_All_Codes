```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fs = pd.read_csv("forestfire.csv")
```

```python
fs.shape
```

```python
fs.head()
```

```python
subset=fs.loc[30:40,['month','temp','rain','wind']]
subset
```

```python
sns.lineplot(subset['wind'])
```

```python
sns.histplot(subset[['temp']],color="blue", kde={'alphabet'})
```

```python
sns.pairplot(fs, diag_kind='kde')
plt.show()
```

```python
subset.boxplot(column=['temp'], grid=True)
plt.title('Temp Box Plot')
```

```python
sns.heatmap(fs.pivot_table(values='temp', index='wind', columns='month'), cmap='RdBu')
plt.show()
```

```python
sns.scatterplot(x=subset['temp'],y=subset['wind'])
plt.show()
```
