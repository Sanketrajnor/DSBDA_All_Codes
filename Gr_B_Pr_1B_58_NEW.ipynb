**Error Correction**

```python
data = {
    'name of person': ['Sanket', 'Abhishek', 'Darshan', 'Sumit', 'Kamlesh'],
    'height': [175, 54, 180, 5.5, 190],
    'unit': ['cm', 'inch', 'cm', 'feet', 'cm']
}

dataf = pd.DataFrame(data)

print("Original DataFrame:\n",dataf)

dataf.loc[dataf['unit'] == 'inch', 'height'] *= 2.54
dataf.loc[dataf['unit'] == 'feet', 'height'] *= 30.48
dataf['unit'] = 'cm'

print("\nCorrected DataFrame:\n",dataf)
```

**Data Model Building**

```python
ht.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix,accuracy_score,classification_report
```

```python
x = ht.drop('output', axis=1)

y = ht['output']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#Decision Tree Classifier
classifier = DecisionTreeClassifier()
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Naive Baye's Classifier
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
