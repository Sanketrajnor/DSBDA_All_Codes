```python
import pandas as pd
import numpy as np
```

```python
fb = pd.read_csv(('facebook.csv'),sep=";")
fb
```

```python
fb.head()
```

```python
fb.shape
```

**Create data subsets**

```python
#Create subsets
subset_1 = fb[['like','share']].loc[:5]
```

```python
subset_2 = fb[['Type','comment']]
subset_2
```

**Sort Data**

```python
#Sort data
subset_1.sort_values(by=['like'],ascending=True)
```

**Transposing** **Data**

```python
#Transposing subset_2
subset_2.transpose()
```

**Merge Data**

```python
#Merging data with vertically
merge_subsets_ver=pd.merge(subset_1,subset_2, left_on='like', right_on='comment')
merge_subsets_ver
```

```python
subset_3=subset_1.loc[22:26]
subset_3
```

```python
subset_4=subset_2.loc[12:16]
subset_4
```

```python
#Concat data with horizontally
merge_subset_hor=pd.concat([subset_4,subset_3])
merge_subset_hor
```

```python
#Joining
join=subset_1.join(subset_2, how='inner', lsuffix='_subset_1', rsuffix='_subset_2')
join
```

**Shape and Reshape Data**

```python
#making subset with rows and columns
subset_5=fb.loc[1:5,['Category','Type','like']]
subset_5
```

```python
pd.melt(subset_5,id_vars=['Category'])
```

```python
res=subset_5.loc[:3]
res
```

```python
pd.pivot(res,columns='Type',values='like')
```
